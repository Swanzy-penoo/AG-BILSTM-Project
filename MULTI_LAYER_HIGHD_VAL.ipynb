```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, precision_recall_curve
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from scipy.spatial.distance import mahalanobis
from scipy.linalg import pinv
from scipy.interpolate import make_interp_spline
from scipy.ndimage import gaussian_filter1d
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, LSTM, Dropout, Bidirectional, Input, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

class CollisionPredictionSystem:
    def __init__(self, data_path, sequence_length=10):
        """
        Initialize the collision prediction system with LSTM

        Args:
            data_path (str): Path to the NGSIM dataset CSV file
            sequence_length (int): Number of time steps to use for LSTM sequences
        """
        self.data_path = data_path
        self.sequence_length = sequence_length
        self.scaler = StandardScaler()
        self.model = None
        self.history = None
        self.feature_names = [
            'v_Vel', 'v_Acc', 'Space_Headway', 'Time_Headway',
            'v_length', 'v_Width', 'Lane_ID'
        ]

    def load_and_preprocess_data(self):
        """
        Load and preprocess the NGSIM dataset (adapted for HighD)
        """
        print("Loading dataset...")
        try:
            self.df = pd.read_csv(self.data_path)
            print(f"Dataset loaded successfully. Shape: {self.df.shape}")
        except FileNotFoundError:
            print(f"Error: Dataset not found at {self.data_path}")
            print("Please ensure the dataset is available at the specified path.")
            return None

        print("\nDataset Info:")
        print(self.df.info())
        print(f"\nFirst few rows:")
        print(self.df.head())

        # Define mapping from HighD dataset columns to expected feature names
        column_mapping = {
            'xVelocity': 'v_Vel',
            'xAcceleration': 'v_Acc',
            'dhw': 'Space_Headway',
            'thw': 'Time_Headway',
            'height': 'v_length', # HighD 'height' typically means vehicle length
            'width': 'v_Width',
            'laneId': 'Lane_ID'
        }

        # Check if all source columns exist in the loaded DataFrame
        missing_source_cols = [col for col in column_mapping.keys() if col not in self.df.columns]
        if missing_source_cols:
            print(f"Error: The following required columns are missing in the dataset: {missing_source_cols}")
            return None

        # Rename columns according to the mapping
        self.df = self.df.rename(columns=column_mapping)
        print("\nColumns renamed to match expected feature names.")

        # Add 'Vehicle_ID' and 'Frame_ID' for compatibility with the rest of the code
        # In HighD, 'id' is the vehicle ID and 'frame' is the frame ID
        if 'id' in self.df.columns:
            self.df['Vehicle_ID'] = self.df['id']
        else:
            print("Error: 'id' column (Vehicle ID) not found in dataset.")
            return None

        if 'frame' in self.df.columns:
            self.df['Frame_ID'] = self.df['frame']
        else:
            print("Error: 'frame' column (Frame ID) not found in dataset.")
            return None


        # The original code had 'v_Class' filtering for automobiles.
        # HighD dataset does not have a 'v_Class' column directly.
        # Assuming the dataset primarily contains relevant vehicle types for this analysis,
        # or that specific filtering logic for HighD is not provided.
        # Thus, removing the 'v_Class' filtering.
        # If specific vehicle types need to be filtered in HighD,
        # new logic based on HighD's meta-data or other columns would be required here.

        # Handle missing values for the *now renamed* feature columns
        self.df = self.df.dropna(subset=self.feature_names)
        print(f"After removing missing values from feature columns: {len(self.df)} records")

        # Create V_Size feature (using renamed columns)
        self.df['V_Size'] = self.df['v_length'] * self.df['v_Width']

        # Recalculate Time_Headway for consistency and to handle cases where v_Vel is zero
        # The original code's intent was to ensure valid Time_Headway values.
        # Using the renamed columns:
        self.df['Time_Headway'] = np.where(
            self.df['v_Vel'] > 0,
            self.df['Space_Headway'] / self.df['v_Vel'],
            0
        )
        print("Recalculated 'Time_Headway' for consistency.")

        # Sort by Vehicle_ID and time (using the newly created 'Vehicle_ID' and 'Frame_ID')
        self.df = self.df.sort_values(['Vehicle_ID', 'Frame_ID'])

        return self.df

    def apply_smoothing(self, window_size=10):
        """
        Apply moving average smoothing to velocity and acceleration data
        """
        print("Applying smoothing filter...")

        for vehicle_id in self.df['Vehicle_ID'].unique():
            vehicle_mask = self.df['Vehicle_ID'] == vehicle_id
            vehicle_data = self.df.loc[vehicle_mask].copy()

            if len(vehicle_data) >= window_size:
                vehicle_data['v_Vel'] = vehicle_data['v_Vel'].rolling(
                    window=window_size, center=True, min_periods=1).mean()
                vehicle_data['v_Acc'] = vehicle_data['v_Acc'].rolling(
                    window=window_size, center=True, min_periods=1).mean()

                self.df.loc[vehicle_mask] = vehicle_data

    def create_collision_labels(self):
        """
        Create collision labels based on acceleration thresholds
        """
        print("Creating collision labels based on acceleration thresholds...")

        acc_mean = self.df['v_Acc'].mean()
        acc_std = self.df['v_Acc'].std()

        print(f"Acceleration - Mean: {acc_mean:.4f}, Std: {acc_std:.4f}")

        acc_percentiles = np.percentile(self.df['v_Acc'], [1, 10, 90, 99])
        print(f"Acceleration percentiles [1%, 10%, 90%, 99%]: {acc_percentiles}")

        level_1_threshold = np.percentile(self.df['v_Acc'], 1)
        level_2_threshold = np.percentile(self.df['v_Acc'], 10)

        print(f"Level 1 threshold (Hard braking): {level_1_threshold:.4f} m/s²")
        print(f"Level 2 threshold (Warning): {level_2_threshold:.4f} m/s²")

        self.df['collision_risk'] = 0

        self.df.loc[self.df['v_Acc'] <= level_2_threshold, 'collision_risk'] = 1
        self.df.loc[self.df['v_Acc'] <= level_1_threshold, 'collision_risk'] = 2

        class_dist = self.df['collision_risk'].value_counts().sort_index()
        print("\nClass distribution:")
        for i, count in enumerate(class_dist):
            percentage = (count / len(self.df)) * 100
            labels = ['Safe', 'Warning', 'Danger']
            print(f"Level {i} ({labels[i]}): {count} ({percentage:.2f}%)")

        return level_1_threshold, level_2_threshold

    def create_sequences(self, data, labels):
        """
        Create sequences for LSTM input

        Args:
            data: Feature data
            labels: Target labels

        Returns:
            X_sequences: 3D array (samples, time_steps, features)
            y_sequences: Labels for each sequence
        """
        print(f"Creating sequences of length {self.sequence_length}...")

        X_sequences = []
        y_sequences = []
        vehicle_ids = self.df['Vehicle_ID'].values

        # Group by vehicle and create sequences
        unique_vehicles = np.unique(vehicle_ids)

        for vehicle_id in unique_vehicles:
            vehicle_mask = vehicle_ids == vehicle_id
            vehicle_indices = np.where(vehicle_mask)[0]

            if len(vehicle_indices) < self.sequence_length:
                continue

            vehicle_data = data[vehicle_indices]
            vehicle_labels = labels[vehicle_indices]

            # Create sequences for this vehicle
            for i in range(len(vehicle_data) - self.sequence_length + 1):
                seq = vehicle_data[i:i + self.sequence_length]
                # Use the label at the end of the sequence (prediction target)
                label = vehicle_labels[i + self.sequence_length - 1]

                X_sequences.append(seq)
                y_sequences.append(label)

        X_sequences = np.array(X_sequences)
        y_sequences = np.array(y_sequences)

        print(f"Created {len(X_sequences)} sequences")
        print(f"Sequence shape: {X_sequences.shape}")

        return X_sequences, y_sequences

    def visualize_class_distribution(self):
        """
        Visualize the class distribution and acceleration distribution
        """
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))

        class_counts = self.df['collision_risk'].value_counts().sort_index()
        labels = ['Safe', 'Warning', 'Danger']
        colors = ['green', 'yellow', 'red']

        axes[0, 0].bar(range(len(class_counts)), class_counts.values, color=colors)
        axes[0, 0].set_xlabel('Risk Level')
        axes[0, 0].set_ylabel('Count')
        axes[0, 0].set_title('Class Distribution (Before Balancing)')
        axes[0, 0].set_xticks(range(len(labels)))
        axes[0, 0].set_xticklabels(labels)

        for i, count in enumerate(class_counts.values):
            axes[0, 0].text(i, count + max(class_counts.values) * 0.01,
                          f'{count}\n({count/len(self.df)*100:.1f}%)',
                          ha='center', va='bottom')

        axes[0, 1].hist(self.df['v_Acc'], bins=50, alpha=0.7, edgecolor='black')
        axes[0, 1].axvline(np.percentile(self.df['v_Acc'], 1), color='red',
                          linestyle='--', label='1st percentile (Danger)')
        axes[0, 1].axvline(np.percentile(self.df['v_Acc'], 10), color='orange',
                          linestyle='--', label='10th percentile (Warning)')
        axes[0, 1].set_xlabel('Acceleration (m/s²)')
        axes[0, 1].set_ylabel('Frequency')
        axes[0, 1].set_title('Acceleration Distribution')
        axes[0, 1].legend()

        sorted_acc = np.sort(self.df['v_Acc'])
        cumulative_freq = np.arange(1, len(sorted_acc) + 1) / len(sorted_acc)

        axes[1, 0].plot(sorted_acc, cumulative_freq)
        axes[1, 0].axvline(np.percentile(self.df['v_Acc'], 1), color='red',
                          linestyle='--', label='1% threshold')
        axes[1, 0].axvline(np.percentile(self.df['v_Acc'], 10), color='orange',
                          linestyle='--', label='10% threshold')
        axes[1, 0].set_xlabel('Acceleration (m/s²)')
        axes[1, 0].set_ylabel('Cumulative Frequency')
        axes[1, 0].set_title('Cumulative Frequency Curve')
        axes[1, 0].legend()

        feature_cols = self.feature_names + ['collision_risk']
        corr_matrix = self.df[feature_cols].corr()

        sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0,
                   ax=axes[1, 1], fmt='.2f')
        axes[1, 1].set_title('Feature Correlation Matrix')

        plt.tight_layout()
        plt.show()

    def genetic_based_oversampling(self, X_minority, y_minority, target_samples):
        """
        Implement genetic-based synthetic oversampling for sequences
        """
        print(f"Generating {target_samples} synthetic sequences...")

        if len(X_minority) < 2:
            return X_minority.copy(), y_minority.copy()

        # Flatten sequences for distance calculation
        X_flat = X_minority.reshape(len(X_minority), -1)

        try:
            cov_matrix = np.cov(X_flat.T)
            if np.linalg.det(cov_matrix) == 0:
                cov_matrix += np.eye(cov_matrix.shape[0]) * 1e-6

            inv_cov = pinv(cov_matrix)
            center = np.mean(X_flat, axis=0)

            distances = []
            for sample in X_flat:
                diff = sample - center
                dist = np.sqrt(diff.T @ inv_cov @ diff)
                distances.append(dist)

        except:
            center = np.mean(X_flat, axis=0)
            distances = [np.linalg.norm(sample - center) for sample in X_flat]

        sorted_indices = np.argsort(distances)[::-1]
        mid_point = len(sorted_indices) // 2

        father_indices = sorted_indices[:mid_point + 1]
        mother_indices = sorted_indices[mid_point + 1:]

        father_set = X_minority[father_indices]
        mother_set = X_minority[mother_indices]

        synthetic_samples = []
        synthetic_labels = []

        samples_generated = 0
        while samples_generated < target_samples:
            if len(father_set) == 0 or len(mother_set) == 0:
                break

            father_idx = samples_generated % len(father_set)
            mother_idx = samples_generated % len(mother_set)

            father = father_set[father_idx]
            mother = mother_set[mother_idx]
            offspring = (father + mother) / 2

            noise = np.random.normal(0, 3, offspring.shape)
            offspring += noise

            synthetic_samples.append(offspring)
            synthetic_labels.append(y_minority[0] if hasattr(y_minority, '__getitem__') else y_minority[0])

            samples_generated += 1

        synthetic_X = np.array(synthetic_samples)
        synthetic_y = np.array(synthetic_labels)

        return synthetic_X, synthetic_y

    def balance_dataset(self):
        """
        Balance the dataset using genetic-based oversampling
        """
        print("Balancing dataset using genetic-based oversampling...")

        X = self.df[self.feature_names].values
        y = self.df['collision_risk'].values

        X_normalized = self.scaler.fit_transform(X)

        # Create sequences
        X_sequences, y_sequences = self.create_sequences(X_normalized, y)

        unique_classes, class_counts = np.unique(y_sequences, return_counts=True)
        max_class_count = max(class_counts)

        print("Original sequence class distribution:")
        for cls, count in zip(unique_classes, class_counts):
            print(f"  Class {cls}: {count} sequences")

        balanced_X = []
        balanced_y = []

        for cls in unique_classes:
            class_mask = y_sequences == cls
            class_X = X_sequences[class_mask]
            class_y = y_sequences[class_mask]

            current_count = len(class_X)
            target_count = max_class_count

            if current_count < target_count:
                samples_needed = target_count - current_count
                synthetic_X, synthetic_y = self.genetic_based_oversampling(
                    class_X, class_y, samples_needed
                )

                combined_X = np.vstack([class_X, synthetic_X])
                combined_y = np.hstack([class_y, synthetic_y])
            else:
                combined_X = class_X
                combined_y = class_y

            balanced_X.append(combined_X)
            balanced_y.append(combined_y)

        self.X_balanced = np.vstack(balanced_X)
        self.y_balanced = np.hstack(balanced_y)

        indices = np.random.permutation(len(self.X_balanced))
        self.X_balanced = self.X_balanced[indices]
        self.y_balanced = self.y_balanced[indices]

        print("\nBalanced sequence class distribution:")
        unique_classes, class_counts = np.unique(self.y_balanced, return_counts=True)
        for cls, count in zip(unique_classes, class_counts):
            print(f"  Class {cls}: {count} sequences")

        print(f"Total balanced sequences: {len(self.X_balanced)}")
        print(f"Sequence shape: {self.X_balanced.shape}")

    def visualize_balanced_distribution(self):
        """
        Visualize the balanced class distribution
        """
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))

        original_counts = np.bincount(self.df['collision_risk'])
        labels = ['Safe', 'Warning', 'Danger']
        colors = ['green', 'yellow', 'red']

        axes[0].bar(range(len(original_counts)), original_counts, color=colors)
        axes[0].set_xlabel('Risk Level')
        axes[0].set_ylabel('Count')
        axes[0].set_title('Original Class Distribution')
        axes[0].set_xticks(range(len(labels)))
        axes[0].set_xticklabels(labels)

        balanced_counts = np.bincount(self.y_balanced.astype(int))

        axes[1].bar(range(len(balanced_counts)), balanced_counts, color=colors)
        axes[1].set_xlabel('Risk Level')
        axes[1].set_ylabel('Count')
        axes[1].set_title('Balanced Class Distribution (Sequences)')
        axes[1].set_xticks(range(len(labels)))
        axes[1].set_xticklabels(labels)

        plt.tight_layout()
        plt.show()

    def create_lstm_model(self, input_shape, num_classes):
        """
        Create an LSTM model for collision prediction

        Args:
            input_shape: Tuple (time_steps, features)
            num_classes: Number of output classes

        Returns:
            Compiled LSTM model
        """
        print("\nBuilding LSTM model...")

        model = Sequential([
            # First Bidirectional LSTM layer
            Bidirectional(LSTM(128, return_sequences=True),
                         input_shape=input_shape),
            Dropout(0.3),
            BatchNormalization(),

            # Second Bidirectional LSTM layer
            Bidirectional(LSTM(64, return_sequences=True)),
            Dropout(0.3),
            BatchNormalization(),

            # Third LSTM layer (unidirectional)
            LSTM(32, return_sequences=False),
            Dropout(0.2),
            BatchNormalization(),

            # Dense layers
            Dense(64, activation='relu'),
            Dropout(0.2),
            Dense(32, activation='relu'),
            Dropout(0.2),

            # Output layer
            Dense(num_classes, activation='softmax')
        ])

        model.compile(
            optimizer=Adam(learning_rate=0.0001),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )

        return model

    def plot_training_curves(self, history, smooth=True, method='spline', sigma=2):
        """
        Plot smooth training and validation accuracy/loss curves with improved styling
        """
        print("\nPlotting training curves...")
        print(f"Smoothing method: {method if smooth else 'none'}")

        epochs = np.array(range(1, len(history.history['accuracy']) + 1))
        train_acc = np.array(history.history['accuracy'])
        val_acc = np.array(history.history['val_accuracy'])
        train_loss = np.array(history.history['loss'])
        val_loss = np.array(history.history['val_loss'])

        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

        def smooth_curve(x_data, y_data, method='spline', sigma=2):
            if method == 'spline' and len(y_data) > 3:
                x_smooth = np.linspace(x_data.min(), x_data.max(), len(x_data) * 10)
                try:
                    spline = make_interp_spline(x_data, y_data, k=min(3, len(x_data)-1))
                    y_smooth = spline(x_smooth)
                    return x_smooth, y_smooth
                except:
                    return x_data, y_data
            elif method == 'gaussian' and len(y_data) > 3:
                y_smooth = gaussian_filter1d(y_data, sigma=sigma)
                return x_data, y_smooth
            return x_data, y_data

        # Plot Accuracy
        if smooth:
            x_train_smooth, train_acc_smooth = smooth_curve(epochs, train_acc * 100, method, sigma)
            x_val_smooth, val_acc_smooth = smooth_curve(epochs, val_acc * 100, method, sigma)

            ax1.plot(x_train_smooth, train_acc_smooth, linewidth=3.0,
                    color='#1f77b4', label='Train Accuracy')
            ax1.plot(x_val_smooth, val_acc_smooth, linewidth=3.0,
                    color='#d62728', label='Val Accuracy')
        else:
            ax1.plot(epochs, train_acc * 100, '-o', linewidth=2, color='#1f77b4',
                    label='Train Accuracy', markersize=4)
            ax1.plot(epochs, val_acc * 100, '-s', linewidth=2, color='#d62728',
                    label='Val Accuracy', markersize=4)

        ax1.set_xlabel('Number of Epochs', fontsize=12, fontweight='normal')
        ax1.set_ylabel('Accuracy (%)', fontsize=12, fontweight='normal')
        ax1.set_title('Multi-Layer BiLSTM Training and Validation Accuracy Curve',
                     fontsize=14, pad=15, fontweight='bold')
        ax1.legend(loc='lower right', fontsize=11, framealpha=0.95, edgecolor='black')
        ax1.grid(True, alpha=0.3, linestyle='--')

        min_acc = min(min(train_acc * 100), min(val_acc * 100))
        max_acc = max(max(train_acc * 100), max(val_acc * 100))
        margin = (max_acc - min_acc) * 0.1
        ax1.set_ylim([max(0, min_acc - margin), min(100.0, max_acc + margin)])

        # Plot Loss
        if smooth:
            x_train_smooth, train_loss_smooth = smooth_curve(epochs, train_loss, method, sigma)
            x_val_smooth, val_loss_smooth = smooth_curve(epochs, val_loss, method, sigma)

            ax2.plot(x_train_smooth, train_loss_smooth, linewidth=3.0,
                    color='#1f77b4', label='Train Loss')
            ax2.plot(x_val_smooth, val_loss_smooth, linewidth=3.0,
                    color='#d62728', label='Val Loss')
        else:
            ax2.plot(epochs, train_loss, '-o', linewidth=2, color='#1f77b4',
                    label='Train Loss', markersize=4)
            ax2.plot(epochs, val_loss, '-s', linewidth=2, color='#d62728',
                    label='Val Loss', markersize=4)

        ax2.set_xlabel('Number of Epochs', fontsize=12, fontweight='normal')
        ax2.set_ylabel('Loss', fontsize=12, fontweight='normal')
        ax2.set_title('Multi-Layer BiLSTM Training and Validation Loss Curve',
                     fontsize=14, pad=15, fontweight='bold')
        ax2.legend(loc='upper right', fontsize=11, framealpha=0.95, edgecolor='black')
        ax2.grid(True, alpha=0.3, linestyle='--')

        plt.tight_layout()
        plt.savefig('lstm_training_curves.png', dpi=300, bbox_inches='tight')
        plt.show()

        print("Training curves saved as 'lstm_training_curves.png'")

    def train_model(self, test_size=0.3, validation_split=0.2, epochs=50):
        """
        Train the LSTM collision prediction model
        """
        print("Training LSTM collision prediction model...")

        X_train, X_test, y_train, y_test = train_test_split(
            self.X_balanced, self.y_balanced,
            test_size=test_size, random_state=42, stratify=self.y_balanced
        )

        y_train_cat = to_categorical(y_train, num_classes=3)
        y_test_cat = to_categorical(y_test, num_classes=3)

        # Create LSTM model
        self.model = self.create_lstm_model(
            input_shape=(X_train.shape[1], X_train.shape[2]),
            num_classes=3
        )

        print(f"\nLSTM Model architecture:")
        self.model.summary()

        early_stopping = EarlyStopping(
            monitor='val_loss', patience=15, restore_best_weights=True, verbose=1
        )

        reduce_lr = ReduceLROnPlateau(
            monitor='val_loss', factor=0.5, patience=7, min_lr=1e-7, verbose=1
        )

        self.history = self.model.fit(
            X_train, y_train_cat,
            validation_split=validation_split,
            epochs=epochs,
            batch_size=64,
            callbacks=[early_stopping, reduce_lr],
            verbose=1
        )

        test_loss, test_accuracy = self.model.evaluate(X_test, y_test_cat, verbose=0)
        print(f"\n{'='*60}")
        print(f"Test Accuracy: {test_accuracy:.4f}")
        print(f"Test Loss: {test_loss:.4f}")
        print(f"{'='*60}")

        y_pred_proba = self.model.predict(X_test)
        y_pred = np.argmax(y_pred_proba, axis=1)

        self.X_test = X_test
        self.y_test = y_test
        self.y_pred = y_pred
        self.y_pred_proba = y_pred_proba

        return self.history

    def evaluate_model(self):
        """
        Comprehensive model evaluation
        """
        print("Evaluating LSTM model performance...")

        labels = ['Safe', 'Warning', 'Danger']
        report = classification_report(
            self.y_test, self.y_pred,
            target_names=labels,
            output_dict=True
        )

        print("\nClassification Report:")
        print(classification_report(self.y_test, self.y_pred, target_names=labels))

        cm = confusion_matrix(self.y_test, self.y_pred)

        plt.figure(figsize=(15, 12))

        # Confusion Matrix
        plt.subplot(2, 3, 1)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                   xticklabels=labels, yticklabels=labels)
        plt.title('Confusion Matrix')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')

        # ROC Curves
        plt.subplot(2, 3, 2)
        for i in range(3):
            y_test_binary = (self.y_test == i).astype(int)
            fpr, tpr, _ = roc_curve(y_test_binary, self.y_pred_proba[:, i])
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, label=f'{labels[i]} (AUC = {roc_auc:.3f})')

        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curves')
        plt.legend()

        # Precision-Recall Curves
        plt.subplot(2, 3, 3)
        for i in range(3):
            y_test_binary = (self.y_test == i).astype(int)
            precision, recall, _ = precision_recall_curve(y_test_binary, self.y_pred_proba[:, i])
            plt.plot(recall, precision, label=f'{labels[i]}')

        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title('Precision-Recall Curves')
        plt.legend()

        # Class-wise metrics
        plt.subplot(2, 3, 4)
        metrics = ['precision', 'recall', 'f1-score']
        class_metrics = []

        class_labels_report_keys = list(report.keys())[:3]

        for metric in metrics:
            class_values = [report[key][metric] for key in class_labels_report_keys]
            class_metrics.append(class_values)

        x = np.arange(len(labels))
        width = 0.25

        for i, metric in enumerate(metrics):
            plt.bar(x + i * width, class_metrics[i], width, label=metric)

        plt.xlabel('Classes')
        plt.ylabel('Score')
        plt.title('Class-wise Performance Metrics')
        plt.xticks(x + width, labels)
        plt.legend()
        plt.ylim([0, 1])

        # Calculate overall metrics FIRST before plotting
        overall_metrics = {
            'Accuracy': accuracy_score(self.y_test, self.y_pred),
            'Macro Avg F1': f1_score(self.y_test, self.y_pred, average='macro'),
            'Weighted Avg F1': f1_score(self.y_test, self.y_pred, average='weighted')
        }

        # Overall metrics
        plt.subplot(2, 3, 5)
        plt.bar(overall_metrics.keys(), overall_metrics.values())
        plt.title('Overall Performance Metrics')
        plt.ylabel('Score')
        plt.ylim([0, 1])
        for i, (key, value) in enumerate(overall_metrics.items()):
            plt.text(i, value + 0.01, f'{value:.3f}', ha='center', va='bottom')

        # Prediction distribution over time
        plt.subplot(2, 3, 6)
        prediction_counts = np.bincount(self.y_pred, minlength=3)
        colors = ['green', 'yellow', 'red']
        plt.bar(range(len(prediction_counts)), prediction_counts, color=colors)
        plt.xlabel('Predicted Risk Level')
        plt.ylabel('Count')
        plt.title('Prediction Distribution')
        plt.xticks(range(len(labels)), labels)

        plt.tight_layout()
        plt.show()

        # Print summary statistics
        print(f"\nLSTM Model Performance Summary:")
        print(f"Overall Accuracy: {overall_metrics['Accuracy']:.4f}")
        print(f"Macro Average F1-Score: {overall_metrics['Macro Avg F1']:.4f}")
        print(f"Weighted Average F1-Score: {overall_metrics['Weighted Avg F1']:.4f}")

        # Calculate and display AUC for each class
        print(f"\nClass-wise AUC Scores:")
        for i in range(3):
            y_test_binary = (self.y_test == i).astype(int)
            fpr, tpr, _ = roc_curve(y_test_binary, self.y_pred_proba[:, i])
            roc_auc = auc(fpr, tpr)
            print(f"{labels[i]}: {roc_auc:.4f}")

    def predict_collision_risk(self, sequence_data):
        """
        Predict collision risk for a new sequence

        Args:
            sequence_data: Array of shape (sequence_length, num_features)
                          Each row contains [v_Vel, v_Acc, Space_Headway,
                          Time_Headway, v_length, v_Width, Lane_ID]

        Returns:
            Predicted class and probabilities
        """
        if self.model is None:
            raise ValueError("Model not trained. Please train the model first.")

        # Ensure correct shape
        if len(sequence_data.shape) == 2:
            # Normalize the sequence
            sequence_normalized = self.scaler.transform(sequence_data)
            # Add batch dimension
            sequence_normalized = sequence_normalized.reshape(1,
                                                             sequence_normalized.shape[0],
                                                             sequence_normalized.shape[1])
        else:
            raise ValueError(f"Expected 2D array, got shape {sequence_data.shape}")

        # Make prediction
        prediction_proba = self.model.predict(sequence_normalized, verbose=0)
        predicted_class = np.argmax(prediction_proba, axis=1)[0]

        labels = ['Safe', 'Warning', 'Danger']

        return {
            'predicted_class': predicted_class,
            'predicted_label': labels[predicted_class],
            'probabilities': {
                labels[i]: float(prediction_proba[0][i]) for i in range(3)
            }
        }

# Main execution function
def main():
    """
    Main function to demonstrate the LSTM collision prediction system
    """
    # Initialize the system with your dataset path
    # Provide the correct path to your NGSIM CSV file
    data_path = "/content/drive/MyDrive/HighD Dataset (total_tracks).csv"

    print("=== LSTM Rear-End Collision Prediction System ===")
    print("Based on: 'A Real-Time Collision Prediction Mechanism With Deep Learning'")
    print("Model: Bidirectional LSTM with Temporal Sequences")
    print("Dataset: Next Generation Simulation (NGSIM)")
    print("=" * 60)

    # Create the prediction system with sequence length of 10 time steps
    sequence_length = 10
    system = CollisionPredictionSystem(data_path, sequence_length=sequence_length)

    # Load and preprocess data
    df = system.load_and_preprocess_data()
    if df is None:
        print("\nPlease update the 'data_path' variable with the correct path to your NGSIM dataset.")
        return

    # Apply smoothing
    system.apply_smoothing()

    # Create collision labels
    system.create_collision_labels()

    # Visualize original data distribution
    print("\nVisualizing original data distribution...")
    system.visualize_class_distribution()

    # Balance the dataset (this will create sequences internally)
    system.balance_dataset()

    # Visualize balanced distribution
    print("\nVisualizing balanced data distribution...")
    system.visualize_balanced_distribution()

    # Train the LSTM model
    history = system.train_model(epochs=50)

    # Plot smooth training curves
    print("\n" + "="*60)
    print("TRAINING CURVES")
    print("="*60)
    system.plot_training_curves(history, smooth=True, method='spline', sigma=2)

    # Evaluate the model
    print("\n" + "="*60)
    print("MODEL EVALUATION")
    print("="*60)
    system.evaluate_model()

    # Example prediction
    print("\n=== Example Prediction ===")
    # Create a sample sequence (10 time steps \u00d7 7 features)
    # Features: [v_Vel, v_Acc, Space_Headway, Time_Headway, v_length, v_Width, Lane_ID]
    sample_sequence = np.array([
        [20.0, -0.5, 35.0, 1.75, 4.5, 1.8, 2],
        [19.5, -1.0, 33.0, 1.69, 4.5, 1.8, 2],
        [18.5, -1.5, 30.0, 1.62, 4.5, 1.8, 2],
        [17.0, -2.0, 27.0, 1.59, 4.5, 1.8, 2],
        [15.0, -2.5, 24.0, 1.60, 4.5, 1.8, 2],
        [12.5, -3.0, 20.0, 1.60, 4.5, 1.8, 2],
        [10.0, -3.5, 17.0, 1.70, 4.5, 1.8, 2],
        [7.5, -3.5, 14.0, 1.87, 4.5, 1.8, 2],
        [5.0, -3.0, 12.0, 2.40, 4.5, 1.8, 2],
        [3.5, -2.5, 10.0, 2.86, 4.5, 1.8, 2]
    ])

    try:
        result = system.predict_collision_risk(sample_sequence)
        print(f"Sample Sequence Shape: {sample_sequence.shape}")
        print(f"Sequence represents: {sequence_length} consecutive time steps")
        print(f"\nPredicted Risk Level: {result['predicted_label']}")
        print(f"Probabilities:")
        for label, prob in result['probabilities'].items():
            print(f"  {label}: {prob:.4f}")
    except Exception as e:
        print(f"Prediction failed: {e}")

    print("\n=== LSTM System Complete ===")
    print("\nKey Advantages of LSTM Model:")
    print("\u2022 Captures temporal dependencies in vehicle trajectories")
    print("\u2022 Learns patterns across multiple time steps")
    print("\u2022 Better suited for time-series prediction")
    print("\u2022 Bidirectional processing considers past and future context")

if __name__ == "__main__":
    main()
```
